# Uniform Distributation

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = np.random.uniform(120, 180 ,100000)
```

```python
plt.hist(data , bins=50 , ec= "black")
plt.show()
```

# Normal Distributation

    In general the density of values around center is always high compare to far regions.
    Probability of picking values around center will higher.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(150 , 20 , 10000)
# mean , std , count

plt.hist(data , bins=50 , ec = "black")
plt.show()
```

```python
print(np.mean(data))
print(np.median(data))
```
