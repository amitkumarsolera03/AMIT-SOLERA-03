# EDA --> Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

## Types of Plots

    Continuous 
            Distribution - Histogram , Line  plot (Continuouse incresing or decresing , Time Series)
            Relationship - Scatter plot , kde plot , rug plot

    Categorical 
            bar chart , pie chart

```python
height = np.random.normal(155 , 10 , 1000).reshape(-1,1)
weight = height*0.5 + np.random.normal(10 , 2 , 1000).reshape(-1,1)

data = pd.DataFrame(np.hstack((height , weight)) , columns = ["height" , "weight"])

data[:5]
```

```python
data["gender"] = np.random.choice(["Male" , "Female"] , 1000)
```

```python
data.head()
```

```python
data["BMI"] = (0.2*data["height"] + 0.3*data["weight"])/2
data.head()
```

## Wide Data 

    no of columns >>> no of Raws
## Long Data 

    no of Raws >>> no of Columns

```python
data["gender"].value_counts()
```

```python
gender = data["gender"].value_counts().reset_index()
gender
```

```python
gender["per"] = gender["count"]/data.shape[0]
gender
```

```python
plt.bar(x = gender["gender"] , height = gender["count"] , color = ["red" , "green"] , alpha=0.5)
plt.show()
```

```python
plt.bar(x = gender["gender"] , height = gender["per"] , color = ["red" , "cyan"])#, alpha=0.5  )#, data= gender)
plt.xlabel("gender")
plt.ylabel("count")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
# sns.countplot(data["gender"])
sns.countplot(x= "gender" , hue = "gender", data = data , palette = "rainbow") 
plt.show()
```

```python
tip = sns.load_dataset("tips")
```

```python
t = tip["day"].value_counts().reset_index()



plt.pie(x= "count" , labels = "day" , data = t , colors = ["r","g","b","c"] , autopct="%0.2f%%" , explode = [0,0,0,0.2] , shadow = True)
plt.show()
plt.bar(x = "day" , height="count" , data=t, color = ["r","g","b","c"])
plt.show()
```

```python
v = np.quantile(data["height"] , [0, 0.25 , 0.5 , 0.75 , 1])

plt.bar(x = ["Min" , "25%" , "50%" , "75%" , "Max"] , height = v)
plt.show()
```
