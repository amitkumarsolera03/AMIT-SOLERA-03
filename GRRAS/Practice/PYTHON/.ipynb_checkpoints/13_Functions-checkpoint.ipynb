```python
def is_prime(num):
    if not (num > 1 and type(num) is int):
        return False

    if num % 2 == 0:
        return False

    if num in [2,3,5,7,11]:
        return True

    for val in range(2, num//2):
        if num % val == 0:
            return False

    return True
```

```python
is_prime(9)
```

```python
# Write a function Square sum of given numbers

def square_sum(nums):
    """
        Function which return Square sum of all given numbers in list nums
    """
    s = 0

    for num in nums:
        s += num**2

    return s
```

```python
nums = [1,2,3,4,5,6]
square_sum(nums)
```

```python
def sq_sum(*nums,sum_ = 0):
    for x in nums:
        sum_ += x**2
    return sum_
```

```python
sq_sum(*range(1,10) , sum_ = 100)
```

# Closures

    A function which contains a function defination inside it and return inner function

```python
def outer(name):
    def inner():
        print("Hello my name is", name)
    return inner
```

```python
func = outer("amit")
```

```python
func()
```

```python
def count():
    x= 0
    def hello():
        nonlocal x
        x+=1
        print(f"{x}.hello world!")
    return hello
```

```python
hi = count()
```

```python
hi()
```

```python
hi()
```

```python
hi()
```

# Decorators 

    special function in python which can add extra fetures to an existing function just like inheritance 

    a function which take function as argument , defines a function inside body , and return the inner function 

```python
def new_hello(old_hello):
    def dec_hello():
        print("_"*20)
        old_hello()
        print("_"*20)
    return dec_hello
```

```python
def hello():
    print("Hello World!!")
```

```python
hello()
```

```python
hi = new_hello(hello)
```

```python
hi()
```

```python
def hello(num):
    if num:
        ret = num + hello(num-1)
        print(num)
    return num
```

```python
def fact(num):
    if num<=1:
        return 1
    return num*fact(num-1)
print(fact(5))
```

```python
def fab(n):
    if n <=1 :
        return 1
    return fab(n-1) + fab(n-2)

print(fab(5))
```
