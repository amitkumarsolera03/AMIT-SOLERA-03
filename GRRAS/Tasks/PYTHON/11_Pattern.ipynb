# 1.	Write Python Code to print these patterns

## (a)

```python
n = int(input("Enter the number of Row's:- "))

i = 0
while i <= n:
    j = 0
    while j < i:
        print(f"{"*":<2}",end="")
        j+=1
    print()
    i+=1
```

## (b)

```python
n = int(input("Enter the number of Row's:- "))

i = 0
while i <= n:
    j = 0
    while j < i:
        print(f"{i:<3}",end="")
        j+=1
    print()
    i+=1
```

## (c)

```python
n = int(input("Enter the number of Row's:- "))

i = 1
while i <= n:
    j = 1
    while j < i:
        print(f"{j:<3}",end="")
        j+=1
    print(i)
    i+=1
```

## (d)

```python
n = int(input("Enter the number of Row's:- "))

i = 1
while i <= n:
    j = 1
    while j < i:
        print(f"{(i+j)-1:<3}",end="")
        j+=1
    print()
    i+=1
```

## (e)

```python
n = int(input("Enter the number of Row's:- "))

i = 0
while i <= n:
    j = 0
    while j < i:
        print(f"{(i+j)%2:<3}",end="")
        j+=1
    print()
    i+=1
```

## (f)

```python
n = int(input("Enter the number of Row's:- "))

i = 1
num = 1
while i <= n:
    j = 1
    while j <= i:
        print(f"{num:<3}",end=" ")
        num+=1
        j+=1
    print()
    i+=1
```

## (g)

```python
n = int(input("Enter the number of Row's:- "))

i = n
while i < 0:
    j = n
    while j < i:
        print(i,end="")
        j-=1
    print()
    i-=1
```
