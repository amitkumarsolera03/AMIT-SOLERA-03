** Import pandas and read in the Ecommerce Purchases csv file and set it to a DataFrame called ecom. **

```python
import pandas as pd
```

```python
ecom = pd.read_csv("Ecommerce Purchases.csv")
```

**Check the head of the DataFrame.**

```python
ecom.head()
```

** How many rows and columns are there? **

```python
r , c = ecom.shape
```

```python
print(f"Total number of Raws are '{r}' and Total number of Column is '{c}'")
```

** What is the average Purchase Price? **

```python
ecom["Purchase Price"].mean()
```

** What were the highest and lowest purchase prices? **

```python
max_price = ecom["Purchase Price"].max()
min_price = ecom["Purchase Price"].min()

print(f"Maximum price is -> {max_price}")
print(f"Minimum price is -> {min_price}")
```

**How many people have English 'en' as their Language of choice on the website?**

```python
lang = ecom['Language'].value_counts()
# ecom.groupby('Language').size()
lang.loc["en"]
```

```python
eng = ecom.groupby('Language').size().loc['en']
print(f"Number of English users: {eng}")
```

** How many people have the job title of "Lawyer" ? **


```python
jobs = ecom["Job"].value_counts()
jobs.loc["Lawyer"]
```

**How many people made the purchase during the AM and how many people made the purchase during PM ?**

**(Hint: Check out [value_counts()](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.value_counts.html) )**

```python
time = ecom["AM or PM"].value_counts()
print(time.loc["AM"])
print(time.loc["PM"])
```

**Someone made a purchase that came from Lot: "90 WT" , what was the Purchase Price for this transaction?**

```python
lot = ecom["Lot"].reset_index().set_index("Lot")
```

```python
index = lot.loc["90 WT"]
index
```

```python
price = ecom["Purchase Price"]
price.iloc[index]
```

** What is the email of the person with the following Credit Card Number: 4926535242672853 **

```python
num = ecom["Credit Card"].reset_index().set_index("Credit Card")
```

```python
index = num.loc[4926535242672853]
print(index)
```

```python
mail = ecom["Email"].iloc[index]
mail
```

** How many people have American Express as their Credit Card Provider *and* made a purchase above $95 ?**

```python
cc = ecom["CC Provider"].reset_index().set_index("CC Provider")
```

```python
import numpy as np
ae = np.array(cc.loc["American Express"])
ae
```

```python
# ae = list(cc.loc["American Express"])
```

```python
x = ecom["Purchase Price"].reset_index()
for i in ae:
    print(x.loc[i])
```
